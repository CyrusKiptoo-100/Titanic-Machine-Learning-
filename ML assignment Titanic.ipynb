**MODEL BUILDING**

```python
titanic.head(5)
```

```python
titanic['Survived'].unique()
```

We use logistics regression because it is in binary form (either 0,1)
0 means false-did not survive
1 means true-Survived

```python
# importing the neccesary libraries for Model Building
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score
```

```python
# we Encode the categorical columns
titanic['Sex'].unique()
```

```python
titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
print(titanic['Sex'].unique())
```

```python
titanic['Embarked'].unique()
```

```python
titanic['Embarked']=titanic['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
print(titanic['Embarked'].unique())
```

```python
x=titanic[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']] # Giving our variables names x and y
y=titanic['Survived']
```

```python
# train and test the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#Give the model variables
model=LogisticRegression(max_iter=1000)

```

```python
# fitting the model
model.fit(x_train,y_train)
```

```python
# PREDICT THE DEPENDENT VARIABLE
y_pred=model.predict(x_test)
```

```python
# CHECKING THE ACCURACY SCORE
accuracy_score(y_test,y_pred)
```

**MODEL TUNING:HYPERPARAMETER**

Since our model had an Accuracy of 0.8044 we need to increase the accuracy to make our model efficient during depoyment

```python
# train the model re train
param_grid={
    'C': [0.1]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
# train  the new data
grid_search.fit(x_train,y_train)
```

```python
# Get best parametre
best_params=grid_search.best_params_
print(best_params)

model.fit(x_train,y_train)
```

```python
#retraining the trained model on the entire training set
best_model=LogisticRegression(**best_params)
best_model.fit(x_train,y_train) #Fit the model
```

```python
# Evaluate the model
y_pred=best_model.predict(x_test)
```

```python
import numpy as np
```

```python
accuracy=accuracy_score(y_test,y_pred)
roc_auc_score =roc_auc_score(y_test,best_model.predict_proba(x_test)[:,1])
print(accuracy)
print(roc_auc_score)
```
